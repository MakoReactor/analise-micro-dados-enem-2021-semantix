```python
import pandas as pd
```

```python
arquivo = "/content/drive/MyDrive/microdados_enem_2021/DADOS/MICRODADOS_ENEM_2021.csv"
```

```python
# Criar um Data Frame com o arquivo.
# Essa etapa pode demorar devido ao tamho do Arquivo.
df = pd.read_csv(arquivo, encoding='ISO-8859-1', sep=';')
```

```python
df.head()
```

```python
# Criar uma lista com os nomes das colunas para facilitar a seleção das colunas
colunas = df.columns.values
colunas
```

```python
# Selecionando algumas colunas específicas para trabalhar com menos dados.
colunas_selecionadas = ['NU_INSCRICAO', 'TP_COR_RACA', 'TP_ESCOLA', 'NU_NOTA_CN','NU_NOTA_CH', 'NU_NOTA_LC','NU_NOTA_MT', 'NU_NOTA_REDACAO', 'Q006']
```

```python
# Criando um DataFrame com apenas as colunas selecionadas 
df_col_selecionadas = df[colunas_selecionadas]
df_col_selecionadas.head()
```

```python
df_col_selecionadas.isna().sum()
```

Neste ponto foi verificado que há muitas NaN (not a number) nas provas o que 
indica abstenção o que é diferente de ter zerado na prova, neste caso o Candidato faltou no dia da prova.

Outra prova que corrobora com essa hipótese é o fato dos valores faltantes serem iguais entre provas minsistradas no mesmo dia.

Desta forma para uma análise de notas é correto retirar dos dados todos os NaN.

- Pode-se concluir que quem zerou na prova de Ciencias Humanas também zerou na prova de Ciências da Natureza, pois as mesmas foram dadas no mesmo dia
- O mesmo pode-se afirmar com as notas de Matemática e Linguagens e códigos e Redação
- Pode se observar que dos 3.389.832  candidados  1.143.988 faltaram no dia da prova de Matemática, Linguagens e Redação (aproximadamente 1/3 de abstenção)

```python
df_col_selecionadas = df_col_selecionadas.dropna()
df_col_selecionadas.isna().sum()
```

## Adicionar uma coluna com informações.
### Para a Coluna Q006 Dicionaŕio de Dados para a Pergunta Q006 = "Qual é a renda mensal de sua família? 
### Essa informação consta no Dicionário de Dados que foi baixado junto com os dados

```python
renda_familiar = {
  'A':	'Nenhuma Renda',
  'B':	'Até R$ 1.100,00',
  'C':	'De R$ 1.100,01 até R$ 1.650,00',
  'D':	'De R$ 1.650,01 até R$ 2.200,00',
  'E':	'De R$ 2.200,01 até R$ 2.750,00',
  'F':	'De R$ 2.750,01 até R$ 3.300,00',
  'G':	'De R$ 3.300,01 até R$ 4.400,00',
  'H':	'De R$ 4.400,01 até R$ 5.500,00',
  'I':	'De R$ 5.500,01 até R$ 6.600,00',
  'J':	'De R$ 6.600,01 até R$ 7.700,00',
  'K':	'De R$ 7.700,01 até R$ 8.800,00',
  'L':	'De R$ 8.800,01 até R$ 9.900,00',
  'M':	'De R$ 9.900,01 até R$ 11.000,00',
  'N':	'De R$ 11.000,01 até R$ 13.200,00',
  'O':	'De R$ 13.200,01 até R$ 16.500,00',
  'P':	'De R$ 16.500,01 até R$ 22.000,00',
  'Q':	'Acima de R$ 22.000,00'
}
```

```python
df_col_selecionadas["RENDA_FAMILIAR"] = df_col_selecionadas['Q006'].map(renda_familiar)
# df_col_selecionadas['RENDA_FAMILIAR'] = [renda_familiar[resp] for resp in df_col_selecionadas['Q006']]
df_col_selecionadas.head()
```

```python
# Obter informações dos dataframe
df_col_selecionadas.info()
```

```python
total = df_col_selecionadas['NU_INSCRICAO'].count()
total
```

### Logo abaixo é demostranda a porcentagem de candidatos que fizeram a prova 
### Agrupados pela Renda Familias.
Percebemos que a maioria dos candidatos é de uma renda relativa baixa, cerca de 
52% têm uma renda familiar até **R$ 2200,00 reais**.

```python
# Porcentagem por Renda Familiar
df_col_selecionadas[['RENDA_FAMILIAR']].value_counts() / total * 100
```

## Com base nas porcentagens de candidatos por faixa de renda familiar surgiu percebi que a maria dos candidatos, cerca de 52%, ou mais da metada, têm uma renda familiar de até R$ 2200,00 reais.

## Com base nessa inferência, surgiu um pergunta:  Seré que a renda familiar tem influência nas notas do ENEM? Candidatos mais pobres tiram notas semelhantes?

## Para testar essa hipótese foram feitos alguns agrupamentos e cálculos. Como visto a seguir.

```python
redacao = df_col_selecionadas[['Q006','RENDA_FAMILIAR', 'NU_NOTA_REDACAO']].groupby(['Q006','RENDA_FAMILIAR']).describe()
redacao
```

```python
matematica = df_col_selecionadas[['Q006','RENDA_FAMILIAR', 'NU_NOTA_MT']].groupby(['Q006','RENDA_FAMILIAR']).describe()
matematica
```

```python
ling_port = df_col_selecionadas[['Q006','RENDA_FAMILIAR', 'NU_NOTA_LC']].groupby(['Q006','RENDA_FAMILIAR']).describe()
ling_port
```

```python
ciencias_natureza = df_col_selecionadas[['Q006','RENDA_FAMILIAR', 'NU_NOTA_CN']].groupby(['Q006','RENDA_FAMILIAR']).describe()
ciencias_natureza
```

```python
ciencias_humanas = df_col_selecionadas[['Q006','RENDA_FAMILIAR', 'NU_NOTA_CH']].groupby(['Q006','RENDA_FAMILIAR']).describe()
ciencias_humanas
```

# Conclusão da Hipótese de Renda x Notas
## O padrão de comportamento de dados foi obeservado de forma semelhante para as notas de Matemática, Linguagens e Códigos (lingua portuguesa), Ciências Naturais, Ciências Humanas e Redação, onde em todos os Casos a média das notas aumentava conforme a renda familiar e o Quartil de 25% da Maior Renda sempre superava o Quartil de 75% da Menor Renda.

## Essa análise demonstra que o fator Renda afeta diretamente a nota dos Candidatos.

## Primeiro foi testada a nota em Redação, foi usado o groupby do pandas e o describe para fazer as contas e de cara já obtive uma boa nuance da resposta que procurava.
- foi observado que tanto candidatos com baixa renda quanto candidatos de alta renda obtiveram notas mínimas igual a ZERO e notas máximas iguais a 1000.
- Também foi notado que o desivo padrão foi menor nos candidatos de renda mais alta portato a variação de notas entre esses candidatos é menor.
- Outro ponto obeservado e preocupante é que o Quartil de 25% dos candidatos de Alta Renda é maior do que Quartil dos candidatos de Baixa Renda.
- E o mais importante, a média da nota de redação aumenta de acordo com a renda.
